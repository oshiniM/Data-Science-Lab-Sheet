```python
import pandas as pd
from apyori import apriori
```

```python
movie_dataset = pd.read_csv("movie_dataset.csv", header=None)
display(movie_dataset.head())
```

```python
movie_dataset.shape
```

```python
records = []
for i in range(0, 7501):
 records.append([str(movie_dataset.values[i, j]) for j in range(0, 20)])
```

```python
records
```

```python
association_rules = apriori(records, min_support=0.01, min_confidence=0.2, 
 min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
association_results
```

```python
association_rules = apriori(records, min_support=0.01, min_confidence=0.5, 
 min_lift=3, min_length=2)
association_results = list(association_rules)
```
